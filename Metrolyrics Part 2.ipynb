```python
#scraping one lyrics page
url = 'http://www.metrolyrics.com/its-the-end-of-the-world-lyrics-rem.html'
response = requests.get(url)
doc = BeautifulSoup(response.text)
doc.prettify()
```

```python
#scraping one
verses = doc.find_all(class_='verse')
list_verses = []
for verse in verses:
    list_verses += [verse.text]
list_verses = "SEPARATOR".join(list_verses)
list_verses
```

```python
def scrape_lyrics(row):
    
    url = row['link']
    print('We are scraping', url)
    response = requests.get(url)
    doc = BeautifulSoup(response.text)
    
    verses = doc.find_all(class_='verse')
    list_verses = []
    for verse in verses:
        row = {}
        list_verses += [verse.text]
    list_verses = "SEPARATOR".join(list_verses)
    row['lyrics'] = list_verses
    return pd.Series(row)    
```

```python
scraped_df = df.apply(scrape_lyrics, axis=1)
```

```python
df = df.merge(scraped_df['lyrics'], left_index=True, right_index=True)
```
