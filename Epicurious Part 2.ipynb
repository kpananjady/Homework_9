```python
#actually writing the function
def scrape_recipe(row):
    url = f"https://www.epicurious.com{row['link']}"
    print('We are now scraping', url)
    response = requests.get(url)
    doc = BeautifulSoup(response.text)

    dict_recipe = {}
    
    list_ingredients = []
    ingredients = (doc.find(class_='ingredients'))
    dict_recipe['ingredient'] = (ingredients.text)

    preparation = doc.find(class_='preparation-steps')
    dict_recipe['steps'] = (preparation.text.strip())
  
    try: 
        tags = doc.find(class_='tags')
        dict_recipe['tags']= (tags.text)
    except:
        pass

    
    return pd.Series(dict_recipe)   
```

```python
scraped_df = df.apply(scrape_recipe, axis=1)
```
