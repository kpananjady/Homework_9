```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
url = 'https://www.goodreads.com/list/show/6934.Science_Fiction_Books_by_Female_Authors'
```

```python
response = requests.get(url)
doc = BeautifulSoup(response.text)
```

```python
doc
```

```python
# Field	Example
# Rank	1
# Title	The Handmaid's Tale
# Author	Margaret Atwood
# Score	score: 30,733
# Votes	314 people voted
# Rating	4.09 avg rating — 1,101,120 ratings
```

```python
doc.find_all(itemtype="http://schema.org/Book")
```

```python
len(doc.find_all(itemtype="http://schema.org/Book"))
```

```python
#details on the page

rows = []
books = doc.find_all(itemtype="http://schema.org/Book")
for book in books:
    row ={}
    row['rank']=(book.find(class_='number').text)
    row['title'] =(book.find(itemprop='name').text)
    row['author']=(book.find(class_='authorName').text)
    row['rating']=(book.find(class_="minirating").text)
    row['score'] =(book.find(class_="smallText uitext").text.split('and')[0].strip()[7:-1])
    row['votes'] = (book.find(class_="smallText uitext").text.split('and')[1].strip())
    rows += [row]
rows
```

```python
df = pd.DataFrame(rows)
```
