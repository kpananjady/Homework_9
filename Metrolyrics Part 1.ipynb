```python
#writing the scraping function 

def scrape_songs():
    list_songs = []
    for page_num in range(0,5): 
        if page_num == 0:
            url = 'http://www.metrolyrics.com/rem-lyrics.html'
        else:    
            url = f'http://www.metrolyrics.com/rem-alpage-{page_num}.html'
        print('We are now scraping', url)
        
        response = requests.get(url)
        doc = BeautifulSoup(response.text)
        tracks = doc.find_all('tr')
        for track in tracks[1:-6]:
            dict_song = {}
            dict_song['name']=(track.find(class_='title').text.strip())
            dict_song['link']=(track.find(class_='title')['href'])
            dict_song['year']=(track.find_all('td')[2].text)
            dict_song['popularity'] = (track.find(class_='bar')['class'][1][7::])
            list_songs += [dict_song]
    return list_songs
```
