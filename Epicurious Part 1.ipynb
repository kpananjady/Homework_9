```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
url = 'https://www.epicurious.com/search/cucumbers'
```

```python
response = requests.get(url)
doc = BeautifulSoup(response.text)
```

```python
doc
```

```python
for card in doc.find_all(class_='recipe-content-card'):
    print(card.prettify())
    print('_________')
```

```python
# Tag/category
# Title
# Summary
# Rating (We'll only want the 2, not the 2 / 4)
# Would make again
# Link/URL
```

```python
rows = []
for card in doc.find_all(class_='recipe-content-card'):
    row = {}
    row['tag'] = (card.find(class_='tag').text)
    row['name'] = (card.find(class_="hed").text.strip())
    row['summary'] = (card.find(class_="dek").text.strip())
    row['rating'] = (card.find(itemprop="ratingValue").text.strip())
    row['make_again'] = (card.find(class_='make-again-percentage').text.strip())
    row['link'] = (card.find('a')['href'])
    rows.append(row)
    
rows    
```

```python
len(rows)
```

```python
#So we have 16 results per page, and 255 results total.That makes almost sixteen pages or so.
rows = []
for page_num in range(1,16):
    
    url = f"https://www.epicurious.com/search/cucumbers?page={page_num}"
    #except:
        #pass
    print("Now scraping", url)
    response = requests.get(url)
    doc = BeautifulSoup(response.text)
    
    for card in doc.find_all(class_='recipe-content-card'):
        row = {}
        row['tag'] = (card.find(class_='tag').text)
        row['name'] = (card.find(class_="hed").text.strip())
        try:
            row['summary'] = (card.find(class_="dek").text.strip())
        except:
            pass
        row['rating'] = (card.find(itemprop="ratingValue").text.strip())
        row['make_again'] = (card.find(class_='make-again-percentage').text.strip())
        row['link'] = (card.find('a')['href'])
        rows.append(row)
    
rows    
```

```python
df = pd.DataFrame(rows)
```
